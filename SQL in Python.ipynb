```python
import psycopg2 as pg
import pandas as pd

conn = pg.connect( host='localhost', user='postgres', password='much17', dbname='latihan_db')

def run(query):
    df = pd.read_sql(query,conn)
    return df
```

## 1. Prepare a list of offices sorted by country, state, city.

```python
run("""select * from _offices
order by country, state, city;""")
```

## 2. How many employees are there in the company?

```python
run("select count(*) from public._employees")
```

## 3. What is the total of payments received?

```python
run("select sum(amount) from public._payments")
```

## 4. Report those payments greater than $100,000.

```python
run("select * from _payments where amount > 100000;")
```

## 5. List the products in each product line.

```python
run("select * from _products order by productline;")
```

## 6. How many products in each product line?

```python
run("select productline, count(productname) from _products group by productline;")
```

## 7. What is the minimum payment received?

```python
run("select min(amount) from _payments;")
```

## 8. Which orders have a value greater than $5,000?
